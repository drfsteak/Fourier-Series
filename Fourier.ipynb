```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
```

```python
t = np.arange(0,1,0.001)
f_clean = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*120*t)
f = f_clean + 2.5 * np.random.randn(len(t))
```

```python
# Graphing Noisy signal with respect to original

plt.figure(figsize=(12, 8))
plt.plot(t, f, label='Noisy Signal')
plt.plot(t, f_clean, label='Original Signal')
plt.legend()
plt.show()
```

```python
# Fourier Transform

filtered_f = fft(f)
amplitude_spectrum = np.abs(filtered_f)


plt.figure(figsize = (12,6))
plt.plot(amplitude_spectrum, label='Amplitude Spectrum')
plt.title('Amplitude Spectrum of Noisy Signal')
plt.xlim(0, 800)
plt.show()
```

```python
cutoff = 300
amplitude_spectrum[amplitude_spectrum < cutoff] = 0

plt.figure(figsize = (12,8))
plt.plot(amplitude_spectrum, label='Filtered FFT')
plt.title('Filtered Fourier Transform')
plt.ylim(0,600)
plt.xlim(0,800)
plt.legend()
```

```python
# Inverse Fourier

inverse = np.fft.ifft(amplitude_spectrum)
plt.figure(figsize=(12, 8))
plt.plot(t,inverse, label='Original Signal')
plt.ylim(-7.5,7.5)
plt.legend()
plt.show()
```
